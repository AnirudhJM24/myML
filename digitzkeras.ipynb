```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
```

```python
(atrain,btrain) , (atest,btest) = keras.datasets.mnist.load_data()
```

```python
len(atrain)
```

```python
plt.matshow(atrain[1])
```

```python
atrainflat = atrain.reshape(len(atrain),28*28)
```

```python
atestflat = atest.reshape(len(atest),784)/255
```

```python
model = keras.Sequential([keras.layers.Dense(100,input_shape=(784,),activation = 'sigmoid'),
                        keras.layers.Dense(10,activation = 'sigmoid')])
```

```python
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = 'accuracy' ,)
```

```python
model.fit(atrainflat,btrain,epochs = 5)
```

```python
prd = model.predict(atestflat)
```

```python
prd[0]
```

```python
for z in prd[:10]:
    print(np.argmax(z))
```
